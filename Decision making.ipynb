1. Find the largest among  two numbers

```python
#integer
num1=int(input("Enter the first number:"))
num2=int(input("Enter the second number:"))

if num1<num2:
  print("The greatest number is",num2)
else:
  print("The greatest number is ",num1)
```

2. Check whether the number is even or odd

```python
#integer
num1=int(input("Enter a number :"))
if num1%2==0:
  print("The number is even",num1)
else:
    print("The number is odd",num1)
```

5. Check the nature of solution of quadratic equations

```python
b=int(input("Enter the value of B:"))
a=int(input("Enter the value of A:"))
c=int(input("Enter the value of C:"))
if b**2-4*a*c==0:
  print("The quadratic equation is equal and real")
elif b**2-4*a*c<0:
  print("The quadratic equation is distinct and imaginary")
else:
  print("The quadratic equation is distinct and real")
```

6. Find the smallest number among three Numbers

7. Find the smallest number in a two digit number

```python

num=int(input("Enter a two digit number :"))
num1=int(num%10)
num2=int((num/10)%10)
if num1>num2:
  print("The smallest number is",num2)
else:
  print("The smallest number is",num1)
```

10. Check whether a year is leap year or not


```python
y=int(input("Enter the year:"))
if y%4==0:
    if y%100==0:
        if y%400==0:
            print("The year is leap year")
        else:
            print("The year is not a leap year")
    else:
        print("The year is a leap year")
else:
    print("The year is not leap year")
```
