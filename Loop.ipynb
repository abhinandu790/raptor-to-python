2. Check whether a number is palindrome or not

```python
num=int(input("Enter the number:"))
r=num
re=0
while (num>0):
      d=num%10
      re=re*10+d
      num=num//10
if r==re:
    print("The given number is palindrome")
else:
    print("The given number is not palindrome")

```
